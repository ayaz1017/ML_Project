```python
X = df1['Message']
Y = df1['Category']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
from sklearn.feature_extraction.text import TfidfVectorizer
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)
X_train_features = feature_extraction.fit_transform(X_train) # fit to the training data first
X_test_features=feature_extraction.transform(X_test) # then transform the test data

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model=LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
prediction_on_test_data=model.predict(X_test_features)
accuracy_score_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
print(f'Accuracy on test data : {accuracy_score_on_test_data}')
```
